Les systèmes d'autocomplétion font désormais partie intégrante des applications modernes. Ils améliorent l'expérience de l'utilisateur en prédisant et en suggérant le mot ou la phrase suivante. Ces systèmes sont largement utilisés dans les moteurs de recherche, les éditeurs de texte, les applications de messagerie, etc. Nous allons nous pencher ici sur le processus de développement d'un système d'autocomplétion à l'aide de modèle N-grams.



**Qu'est ce qu'un système d'autocomplétion ?**

---

Un système d'autocomplétion est un système conçu pour prédire et suggérer le mot ou la phrase suivante la plus probable en fonction des données entrées par l'utilisateur. Pour ce faire, il analyse de grands corpus de textes et utilise des techniques statistiques et d'apprentissage automatique.



**Comment marche le modèle à N-grams?**



Les modèles de langage à n-grammes sont une technique fondamentale utilisée dans les systèmes d'autocomplétion pour prédire le mot suivant sur la base des mots précédents ou du contexte. Ils analysent les fréquences des séquences de mot dans un corpus de texte donné et calculent la probabilité d'un mot en fonction de son contexte (N-1 mots précédents).

En effet,les n-grammes sont des séquences contiguës de « n » éléments. Ces éléments peuvent être des caractères, des mots ou même des syllabes, selon la granularité souhaitée. La valeur de « n » détermine l'ordre du N-gramme.

# Exemple:
Considérons que nos éléments de base sont des mots et n= 2, quel sera les 2-grammes de la phrase suivante: « Linkedin est un réseau social professionnel » ?

 Les bigrammes (2-grammes) seraient [« Linkedin est »], [« est un »], [« un réseau »], [« reseau social »] et [« social professionnel »].



```python
from nltk import ngrams
sentence = 'Linkedin est un réseau social professionnel.'
n = 2
bigrams = ngrams(sentence.split(), n)
for grams in bigrams:
  print (grams)
```



# Maintenant à quoi tout cela sert et comment cela est utilisé pour la prédiction du mot suivant dans un système d'auto complétion ?

Supposons que je donne au système la phrase « Merci beaucoup pour votre » et que j'attends du système qu'il prédise le mot suivant. Nous savons tous les deux que le mot suivant est « aide » avec une très forte probabilité. Mais comment le système le saura-t-il et pourra me faire la bonne proposition ?


# Exemple:

Supposons que nous travaillions avec un modèle à bigrammes et que nous ayons les phrases suivantes comme données  d'entraînement :


*   Merci beaucoup pour votre aide.

*  J'apprécie vraiment votre aide.

*   Excusez-moi, savez-vous quelle chemin prendre ?

*  Je suis vraiment désolé de ne pas vous avoir appelé.

*  J'aime beaucoup votre chaussure.

S'il s'agit d'un modèle à bigrammes, le  modèle va apprendre l'occurrence d'un mot sur deux  afin de déterminer la probabilité qu'un mot apparaisse après un autre. Par exemple, d'après les premières, deuxième et cinquième phrases de l'exemple ci-dessus, nous savons qu'après le mot « votre», nous pouvons voir apparaître soit le mot « aide », soit le mot « chaussure ». Le modèle calculera donc la probabilité de chacune de ces séquences en se basant sur les données précédentes (les phrases d'entrainement) de cette façon:


Nombre de  la séquence (votre aide) / Nombre du mot (votre)
= 2 / 3
= 0,66

De même, pour la seconde possibilité:

Nombre de  la séquence (votre chaussure) / Nombre du mot (votre)
= 1 / 3
= 0,33


Ainsi, lorsque je taperai la phrase « Merci beaucoup pour votre » et que je m'attends à ce que le modèle me suggère le mot suivant(sachant que je désire ecrire « Merci beaucoup pour votre aide»  ), il n'obtiendra la bonne réponse  deux fois sur trois, car la probabilité de la bonne réponse n'est que de 2/3. Le modèle me suggèrera donc [« aide »]  deux fois sur trois et  [« chaussure »] une fois sur trois.
